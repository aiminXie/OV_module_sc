```r
proliferationGene <- c('AURKA', 'BUB1', 'CCNB1', 'CCND1', 'CCNE1', 'DEK', 'E2F1', 'FEN1', 'FOXM1', 'H2AFZ', 
                       'HMGB2', 'MCM2', 'MCM3','MCM4', 'MCM5', 'MCM6', 'MKI67', 'MYBL2', 'PCNA', 'PLK1', 'TOP2A', 'TYMS','ZWINT')
modules_list$proliferation <- proliferationGene
```

```r
library(GSVA)
```

```r
modules_score2 <- gsva(expr = OV_TPM_log,gset.idx.list = modules_list,method='gsva',kcdf="Gaussian")
```

```r
colnames(modules_score2) <- gsub(pattern = '\\.',replacement = '-',colnames(modules_score2))
colnames(modules_score2) <- substr(colnames(modules_score2),start = 1,stop = 12)
modules_score2[1:6,1:6]
```

```r
load(file = './data/bulk/dataMerge_subset.RData')
```

```r
bothSample <- intersect(colnames(modules_score2),dataMerge_subset$patient_id)

dataMerge_subset2 <- dataMerge_subset[match(bothSample,dataMerge_subset$patient_id),]
dataMerge_subset2$HRD_status <- ifelse(dataMerge_subset2$HRD>=41,'HRD','HRP')
modules_score2 <- modules_score2[,bothSample]
tmp_data <- dataMerge_subset2
```

```r
tmp_data$M53<- modules_score2['c1_53',]
tmp_data$M769<- modules_score2['c1_769',]
tmp_data$M7<- modules_score2['c1_7',]
tmp_data$proliferation<- modules_score2['proliferation',]
```

```r
head(tmp_data)
```

```r
library(survminer)
require("survival")
library("ggsci")
library(gridExtra)
```

```r
tmp_data$HRD_status  <- ifelse(tmp_data$HRD_status=='HRD','HRD+','HRD-')
```

```r
tmp_data$M53.group <- ifelse(tmp_data$M53>=median(tmp_data$M53),'ROS+','ROS-') 
tmp_data$M769.group <- ifelse(tmp_data$M769>=median(tmp_data$M769),'PRO+','PRO-')
tmp_data$Combined.group <- paste0(tmp_data$M53.group,tmp_data$HRD_status)
```

```r
table(subset(tmp_data,!is.na(HRD_status))$Combined.group)
```

```r
labs <- c(paste0('ROS-HRD-(n=',table(subset(tmp_data,!is.na(HRD_status))$Combined.group)['ROS-HRD-'],')'),
          paste0('ROS-HRD+(n=',table(subset(tmp_data,!is.na(HRD_status))$Combined.group)['ROS-HRD+'],')'),
          paste0('ROS+HRD-(n=',table(subset(tmp_data,!is.na(HRD_status))$Combined.group)['ROS+HRD-'],')'),
          paste0('ROS+HRD+(n=',table(subset(tmp_data,!is.na(HRD_status))$Combined.group)['ROS+HRD+'],')')
         )
fit <- survfit(Surv(FFI.time, FFI.event) ~ Combined.group, data = subset(tmp_data,!is.na(HRD_status)))
OS_surv <- ggsurvplot(
  fit, 
  data = subset(tmp_data,!is.na(HRD_status)),
  title='TCGA PFS',
  size = 1,                 # change line size
  conf.int = F,          # Add confidence interval
    pval.method = TRUE,
  pval = TRUE,              # Add p-value
  risk.table = T,        # Add risk table
    palette =  'jco',
  risk.table.height = 0.3,
  xlab ='Months',
 legend=c(0.7,0.7),
    pval.coord=c(70,0.3),
    pval.method.coord=c(70,0.4),
 legend.labs = labs,
 legend.title='Combined Group',
  ggtheme = theme_bw()      # Change ggplot2 theme
)
options(repr.plot.height = 5, repr.plot.width = 5)
OS_surv
```

## 2. The association of ROS and proliferation GSE133296 for Chemotherapy samples

```r
load(file = './data/bulk/GEO/treatment/GSE15622.RData')
load(file = './data/bulk/GEO/treatment/GSE28739.RData')
```

```r
modules_list$c1_53
```

```r
library(GSVA)
moduleScore <- gsva(expr=GSE15622$exp, gset.idx.list=modules_list, kcdf="Gaussian",method='gsva', parallel.sz=1)
```

```r
library(ggplot2)
library(ggpubr)
metaData <- GSE15622$annotation
dim(metaData)
metaData$M53 <- moduleScore['c1_53',]
metaData$proliferation <- moduleScore['proliferation',]
metaData$M7 <- moduleScore['c1_7',]


metaData <- subset(metaData,Patient%in%names(table(metaData$Patient))[table(metaData$Patient)>1])
metaData <- do.call(rbind,split(metaData,f = metaData$Patient))
metaData <- rbind(subset(metaData,treatState=='post-treatment'),subset(metaData,treatState=='pre-treatment'))

compare_means(M53 ~ treatState, data = metaData)

p1 <- ggpaired(metaData, x = "treatState", y = "M53",xlab = NULL,
         ylab = "M53 GSVA",show.legend = TRUE ,title = 'GSE15622',
         color = "treatState", line.color = "gray", line.size = 0.4,
         palette = "jco")+
  stat_compare_means(paired = TRUE)
options(repr.plot.height = 5, repr.plot.width = 4)
p1
```

```r
head(metaData)
```

```r
library(ggplot2)
library(ggpubr)
compare_means(proliferation ~ treatState, data = metaData)

p2 <- ggpaired(metaData, x = "treatState", y = "proliferation",xlab = NULL,
         ylab = "Cell Cycle",show.legend = TRUE ,title = 'GSE15622',
         color = "treatState", line.color = "gray", line.size = 0.4,
         palette = "jco")+
  stat_compare_means(paired = TRUE)+coord_flip()
p2
```

```r
test <- ggscatterhist(
  metaData, x = "proliferation", y = "M53",add = "reg.line",conf.int = TRUE,cor.coef = T,add.params = list(color = "red"),
    palette = "jco",cor.coeff.args = list(method = "pearson", label.sep = "\n"),
  color = "treatState",size = 3, alpha = 0.6,
  margin.plot = "boxplot",margin.params = list(color = "treatState"),main.plot.size = 3, margin.plot.size = 3,
  ggtheme = theme_bw()
)
test$yplot <- p1
test$xplot <- p2
print(test, margin.space = F, main.plot.size = 3, 
            margin.plot.size = 3)
```

```r
pdf(file = './output/bulk/section_5/1.cellCycleVSM53_GSE15622_Chemotherapy.pdf',width = 4,height = 4)
print(test, margin.space = F, main.plot.size = 3, 
            margin.plot.size = 3)
dev.off()
```
