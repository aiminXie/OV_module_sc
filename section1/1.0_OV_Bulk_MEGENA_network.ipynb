# 1.MEGENA co-expression network analysis

```r
OV_TPM <- load(file = './data/bulk/OV_TPM.RData')
#load(file = './data/bulk/OV_TPM.RData')
```

```r
##gene ID convert for Exp matirx
OV_TPM <- exp_idchange(data = OV_TPM,idRaw='ENSEMBL',idTo='SYMBOL')
dim(OV_TPM)
```

```r
OV_TPM_log <-  log2(OV_TPM+1)
```

```r
saveRDS(OV_TPM_log,file = './data/bulk/TCGA_OV_TPM_log2.RDS')
```

```r
# input parameters
n.cores <- 10; ## number of cores
doPar <- TRUE; ## parallelize? 
method = "pearson" ## correlation method
FDR.cutoff = 0.05 ## correlation significance cutoff
module.pval = 0.05 ## module p-value
hub.pval = 0.05 ## hub p-value
```

```r
datExpr <- OV_TPM_log
```

```r
datExpr <- datExpr[!grepl(pattern = '^RP[LS]',rownames(datExpr)),]
dim(datExpr)
```

```r
## Extract highly variable genes
dim(datExpr)
m.mad <- apply(datExpr,1,mad,na.rm=T)
dataExprVar <- datExpr[which(m.mad > 
                                max(quantile(m.mad, probs=seq(0, 1, 0.1))[6],0.01)),]
dim(dataExprVar)
```

```r
library(parallel)
```

```r
#####
# calculate correlation
ijw <- calculate.correlation(datExpr = dataExprVar,doPerm = 10,method = method,FDR.cutoff = FDR.cutoff)
```

```r
library(doParallel)
```

```r
library(doParallel)
cl <- parallel::makeCluster(n.cores)
registerDoParallel(cl)
clusterEvalQ(cl, .libPaths(c("/home/data/refdir/Rlib","/home/data/t060324/R/x86_64-pc-linux-gnu-library/4.1")))
```

```r
## grow a Planar Filtered Network (PFN) using the PMFG algorithm
el <- calculate.PFN(ijw[,1:3],doPar = doPar,num.cores = n.cores)
```

```r
## export result
saveRDS(el,file = "./output/MEGENA/TCGA_MEGENA_Network.RDS")
```

```r
#el <- readRDS(file = "./output/MEGENA/TCGA_MEGENA_Network.RDS")
```

```r
write.table(el,file = "./output/MEGENA/TCGA_MEGENA_Network.txt",sep = "\t",row.names = F,col.names = T,quote = F)
rm(ijw)
```

```r
# Multiscale Clustering Analysis (MCA) to identify co-expression modules
g <- graph.data.frame(el,directed = F)

TCGA_MEGENA.output <- do.MEGENA(g,
 mod.pval = module.pval,hub.pval = hub.pval,remove.unsig = TRUE,
 min.size = 10,max.size = vcount(g)/4,
 doPar = TRUE,num.cores = n.cores,n.perm = 100,
 save.output = TRUE)
```

# 2.Extraction of prognostic gene signatures

```r
load(file = './data/bulk/dataMerge_subset.RData')
```

```r
OV_TPM_log <- readRDS(file = './data/bulk/TCGA_OV_TPM_log2.RDS')
```

```r
OV_TPM_log[1:6,1:6]
```

```r
dataExprVar <- OV_TPM_log
colnames(dataExprVar) <- substr(colnames(dataExprVar),start = 1,stop = 12)
colnames(dataExprVar) <- gsub(pattern = '\\.',replacement = '-',colnames(dataExprVar))
rownames(dataMerge_subset) <- dataMerge_subset$patient_id
bothSample <- intersect(colnames(dataExprVar),dataMerge_subset$patient_id)
dataMerge_subset2 <- dataMerge_subset[bothSample,]
dataExprVar <- dataExprVar[,bothSample]
```

#### prognostic gene signatures(OS)

```r
expr_M <- dataExprVar
tmp_data <- dataMerge_subset2

OS_COX <- lapply(rownames(expr_M),function(gene1){
    tmp_data$geneExp <- expr_M[gene1,]
    res.cox_OS <- coxph(Surv(OS.time, OS.event) ~ geneExp, data = tmp_data)
    res <- summary(res.cox_OS)
    return(data.frame(gene=gene1,z=res$coefficients[1,4],p=res$coefficients[1,5]))
})
```

```r
OSCox_res <- do.call(rbind,OS_COX)
OSCox_res<- OSCox_res[order(OSCox_res$z,decreasing = T),]
OSCox_res <- na.omit(OSCox_res)
```

#### prognostic gene signatures(PFS)

```r
expr_M <- dataExprVar
tmp_data <- dataMerge_subset2
FFI_COX <- lapply(rownames(expr_M),function(gene1){
    tmp_data$geneExp <- expr_M[gene1,]
    res.cox_OS <- coxph(Surv(FFI.time, FFI.event) ~ geneExp, data = tmp_data)
    res <- summary(res.cox_OS)
    return(data.frame(gene=gene1,z=res$coefficients[1,4],p=res$coefficients[1,5]))
})
FFICox_res <- do.call(rbind,FFI_COX)
head(FFICox_res)
```

```r
FFICox_res<- FFICox_res[order(FFICox_res$z,decreasing = T),]
FFICox_res <- na.omit(FFICox_res)
head(FFICox_res,20)
tail(FFICox_res,20)
```

#### prognostic gene signatures(DFS)

```r
expr_M <- dataExprVar
tmp_data <- dataMerge_subset2
DFS_COX <- lapply(rownames(expr_M),function(gene1){
    tmp_data$geneExp <- expr_M[gene1,]
    res.cox_OS <- coxph(Surv(DFS.time, DFS.event) ~ geneExp, data = tmp_data)
    res <- summary(res.cox_OS)
    return(data.frame(gene=gene1,z=res$coefficients[1,4],p=res$coefficients[1,5]))
})
DFSCox_res <- do.call(rbind,DFS_COX)
head(DFSCox_res)
```

```r
DFSCox_res<- DFSCox_res[order(DFSCox_res$z,decreasing = T),]
DFSCox_res <- na.omit(DFSCox_res)
head(DFSCox_res,20)
tail(DFSCox_res,20)
```

```r
TCGA_COXres <- list(OSCox_res=OSCox_res,FFICox_res=FFICox_res,DFSCox_res=DFSCox_res)
```

#### platinum response gene signature

```r
expr_M <- dataExprVar
tmp_data <- dataMerge_subset2
ResistantIndex <- which(tmp_data$platinum_response=='Resistant')
SensitiveIndex <- which(tmp_data$platinum_response=='Sensitive')
```

```r
diffGene_platinum_response <- sapply(rownames(expr_M),function(x){
  Resistant <- expr_M[x,ResistantIndex]
  Sensitive <- expr_M[x,SensitiveIndex]
  lgfc <- mean(Resistant)-mean(Sensitive)
  if(lgfc>=0){
    res <- t.test(Resistant,Sensitive, paired = F,alternative = 'greater')
  }else{
    res <- t.test(Resistant,Sensitive, paired = F,alternative = 'less')
  }
  return(data.frame(gene=x,logFC=lgfc,t=res$statistic,p_value=res$p.value))
},simplify = F)
diffGene_platinum_response <- do.call(rbind,diffGene_platinum_response)
diffGene_platinum_response <- diffGene_platinum_response[order(diffGene_platinum_response$logFC),]
diffGene_platinum_response <- na.omit(diffGene_platinum_response)
```

```r
module2hub_list <- list()
for(i in rownames(module_membership_fisher)){
    hub_gene <- module_membership_fisher[i,4]
    module2hub_list[[i]] <- strsplit(x = gsub(pattern = '\\([0-9]+\\)',replacement = '',hub_gene),split = ',')[[1]]
}
```

```r
head(TCGA_MEGENA.output$node.summary)
```

```r
all_node <- unique(do.call(c,summary.TCGA_MEGENA.output$modules[all_signiff_modules_fisher]))
ALLnode_select_TCGA_MEGENA <- lapply(all_node,function(node){
   node.summary_TMP <-  TCGA_MEGENA.output$node.summary[node,]
   module_ship <- strsplit(x = node.summary_TMP[1,2],split = ',')[[1]]
    if(length(module_ship)==1){
        module <- module_ship
        module2 <- module_ship
    }else{
        #module_ship <- rev.Vector(module_ship)
        for(j in rev(module_ship)){
            if(j %in% all_signiff_modules_fisher){
                module <- j
                break
            }
        }
        for(j in module_ship){
            if(j %in% all_signiff_modules_fisher){
                module2 <- j
                break
            }
        }
    }
    degree <- node.summary_TMP[1,4]
    ishub <- ifelse(node %in% module2hub_list[[module]],1,0)
    return(data.frame(node=node,mainModule=module2,module=module,degree=degree,ishub=ishub))
})
```

```r
ALLnode_select_TCGA_MEGENA <- do.call(rbind,ALLnode_select_TCGA_MEGENA)
head(ALLnode_select_TCGA_MEGENA)
```
