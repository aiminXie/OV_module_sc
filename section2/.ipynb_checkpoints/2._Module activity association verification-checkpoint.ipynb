```r
GEO_moduleCor <- lapply(GEO_genesetssGSEA,function(x){cor(t(x))})
```

```r
GEO_moduleCor$GSE28739
```

```r
GEO_moduleCor_count <- lapply(GEO_moduleCor,function(x){
    ifelse(x>0.4,1,ifelse(x< -0.4,-1,0))
})
```

```r
rep(0,9)
```

```r
GEO_moduleCor_count$GSE28739 <- cbind(GEO_moduleCor_count$GSE28739,c_10=rep(0,9))
GEO_moduleCor_count$GSE28739 <- rbind(GEO_moduleCor_count$GSE28739,c_10=c(rep(0,9),1))
GEO_moduleCor_count$GSE28739
```

```r
moduleCor_count_pos <- GEO_moduleCor_count[[1]]
moduleCor_count_pos[which(moduleCor_count_pos<0)]=0
for(i in GEO_moduleCor_count[-1]){
    i[which(i<0)]=0
    moduleCor_count_pos <- moduleCor_count_pos+i
}
```

```r
moduleCor_count_neg <- GEO_moduleCor_count[[1]]
moduleCor_count_neg[which(moduleCor_count_neg>0)]=0
for(i in GEO_moduleCor_count[-1]){
    i[which(i>0)]=0
    moduleCor_count_neg <- moduleCor_count_neg+i
}
```

```r
colnames(moduleCor_count_pos) <- rownames(moduleCor_count_pos) <- sub(pattern = 'c1_',replacement = 'M',rownames(moduleCor_count_pos))
colnames(moduleCor_count_neg) <- rownames(moduleCor_count_neg) <- sub(pattern = 'c1_',replacement = 'M',rownames(moduleCor_count_neg))
```

```r
library(corrplot)
```

```r
moduleCor_count_pos
moduleCor_count_neg
```

```r
moduleCor_count <- moduleCor_count_pos+moduleCor_count_neg
moduleCor_count
```

### Plot the scatter plot of module correlation

```r
moduleGSVA_38 <- lapply(GEO_genesetssGSEA,function(x){
    data.frame(M3=x['c1_3',],M8=x['c1_8',])
})
```

```r
moduleGSVA_753 <- lapply(GEO_genesetssGSEA,function(x){
    data.frame(M7=x['c1_7',],M53=x['c1_53',])
})
```

```r
library(ggpubr)
```

```r
scatterPlot_list <- list()
for(i in names(moduleGSVA_38)){
    scatterPlot_list[[i]] <- ggscatter(data = moduleGSVA_38[[i]],x = 'M3',y = 'M8',color = 'gray',size = 0.5, add = "reg.line", conf.int = TRUE,title = i,add.params = list(color = "red"),cor.coeff.args = list(method = "pearson", label.sep = "\n",color='black'),
          cor.coef = TRUE)
}
```

```r
scatterPlot_list2 <- list()
for(i in names(moduleGSVA_753)){
    scatterPlot_list2[[i]] <- ggscatter(data = moduleGSVA_753[[i]],x = 'M7',y = 'M53',color = 'gray',size = 0.3, add = "reg.line", conf.int = TRUE,title = i,fill = 'grey',add.params = list(color = "red"),cor.coeff.args = list(method = "pearson", label.sep = "\n",color='black'),
          cor.coef = TRUE)
}
```

```r
library(gridExtra)
```

```r
options(repr.plot.height = 7, repr.plot.width = 10)
ggarrange(plotlist = scatterPlot_list,ncol=5,nrow = 3)
```

```r
options(repr.plot.height = 7, repr.plot.width = 10)
ggarrange(plotlist = scatterPlot_list2,ncol=5,nrow = 3)
```

```r
pdf(file = './output/bulk/section_2/4.scatterPlot_c3_c8.pdf',width = 10,height = 7)
ggarrange(plotlist = scatterPlot_list,ncol=5,nrow = 3)
dev.off()
```

```r
pdf(file = './output/bulk/section_2/4.scatterPlot_c7_c53.pdf',width = 10,height = 7)
ggarrange(plotlist = scatterPlot_list2,ncol=5,nrow = 3)
dev.off()
```
